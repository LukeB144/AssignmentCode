```python
import math

# Define constants for color formatting
RED = "\033[31m"
BLUE = "\033[34m"
RESET = "\033[0m"

def fact(n):
    return math.factorial(n)

def bin(n, r):
    return int(fact(n) / (fact(n - r) * fact(r)))

def col(n):
    colours = [RED, BLUE]
    return f"{colours[n]}{n}{RESET}"

rows = 60  # Adjust the number of rows

for n in range(rows):
    row_print = " " * (rows - n)  # Add leading spaces for alignment
    for r in range(n + 1):
        row_print += f"{col(bin(n, r) % 2)} "  # Add each number with a space
    print(row_print)
```
